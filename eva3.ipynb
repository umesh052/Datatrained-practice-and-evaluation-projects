```python
#for balancing dataset
from sklearn.utils import resample
```

```python
z=df[df.Churn==0]
a=df[df.Churn==1]
```

```python
up=resample(a,replace=True,n_samples=len(z),random_state=0)
```

```python
new_df=pd.concat([z,up])
new_df
```

```python
new_df.Churn.value_counts().plot.bar(color='green')
```

```python
x=new_df.drop(columns=["Churn","customerID","gender","SeniorCitizen"],axis=1)
y=new_df["Churn"]
```

# model building

```python
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score,KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
from sklearn.ensemble import GradientBoostingClassifier,BaggingClassifier,RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.decomposition import PCA
from xgboost import XGBClassifier
```

```python
#scaling dataset
scale=StandardScaler()
x_scaled=scale.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.25,random_state=150)
```

# RandomForestClassifier

```python
clf = RandomForestClassifier()
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
print("\nClassification report :\n",classification_report(y_test,y_pred))
```

```python
accuracy_score(y_test,y_pred)
```

```python
cross_val_score(clf,x_scaled,y,cv=5).mean()
```

```python
rf_conf_mat = confusion_matrix(y_test,y_pred)
print('confusion_matrix ===>\n',rf_conf_mat)
```
