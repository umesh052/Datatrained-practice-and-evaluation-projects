```python
x=df.drop(columns='Happiness Score')
y=df['Happiness Score']
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
```

```python
scale=StandardScaler()
x_scaled=scale.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.25,random_state=50)
```

```python
clf = DecisionTreeRegressor()
clf.fit(x_train,y_train)
```

```python
clf.score(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
cross_val_score(clf,x_scaled,y,cv=3).mean()
```

```python
linear = LinearRegression()
linear.fit(x_train,y_train)
```

```python
linear.score(x_train,y_train)
```

```python
y_pred=linear.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
cross_val_score(linear,x_scaled,y,cv=3,).mean()
```

```python
gd = GradientBoostingRegressor()
gd.fit(x_train,y_train)
```

```python
y_pred=gd.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
svr=SVR()
svr.fit(x_train,y_train)
```

```python
svr.score(x_train,y_train)
```

```python
y_pred=svr.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
cross_val_score(svr,x_scaled,y,cv=5,).mean()
```

```python
param_search ={"C":[1,5,10,15,20],"kernel":["linear","poly","rbf","sigmoid"],"degree":[1,2,3]}
```

```python
grid_search =GridSearchCV(SVR(),param_grid=param_search)
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
svr = SVR(C=1,kernel="poly",degree=1)
```

```python
svr.fit(x_train,y_train)
```

```python
y_pred=svr.predict(x_test)
```

```python
cross_val_score(svr,x_scaled,y,cv=3).mean()
```
